```python
# Creating a 'loyalty_flag' based on the max order number for each user
ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform('max')

# Assigning loyalty categories using loc()
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'

# Checking the distribution of loyalty categories
ords_prods_merge['loyalty_flag'].value_counts()
```

# Step 6

```python
# Calculate the average price for each user
avg_price_per_user = ords_prods_merge.groupby('user_id')['prices'].mean()
```

```python
# Create a new column 'spending_flag' using loc based on the average price
ords_prods_merge['spending_flag'] = 'Low spender'  # Default to 'Low spender'
ords_prods_merge.loc[ords_prods_merge['user_id'].isin(avg_price_per_user[avg_price_per_user >= 10].index), 'spending_flag'] = 'High spender'
# Checking the distribution
ords_prods_merge['spending_flag'].value_counts()
```

# Step 7

```python
# Calculate the median of 'days_since_prior_order' for each user
median_days_per_user = ords_prods_merge.groupby('user_id')['days_since_prior_order'].median()
```

```python
# Create a new column 'order_frequency_flag' using loc based on the median value
ords_prods_merge['order_frequency_flag'] = 'Non-frequent customer'  # Default to 'Non-frequent customer'
# Update the flag based on the median value
ords_prods_merge.loc[ords_prods_merge['user_id'].isin(median_days_per_user[median_days_per_user <= 10].index), 'order_frequency_flag'] = 'Frequent customer'
ords_prods_merge.loc[ords_prods_merge['user_id'].isin(median_days_per_user[(median_days_per_user > 10) & (median_days_per_user <= 20)].index), 'order_frequency_flag'] = 'Regular customer'
# Checking the distribution
ords_prods_merge['order_frequency_flag'].value_counts()
```
